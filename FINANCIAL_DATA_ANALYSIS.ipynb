<h4> What is the overall sales trend

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')  # Corrected format

# Calculate the minimum date value
df['Date'].min()
```

```python
df['Date'].max()
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Group by 'Month Name' and 'Year', and sum '  Sales '
df_trend = df.groupby(['Year',' Month Name '])['  Sales '].sum().reset_index()

# Print the result
print(df_trend)
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Group by 'Month Name' and 'Year', and sum '  Sales '
df_trend = df.groupby(['Year', ' Month Name '])['  Sales '].sum().reset_index()

# Create a custom sort order that combines 'Year' and ' Month Name '
month_order = ['September', 'October', 'November', 'December', 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August']

# Strip extra spaces from ' Month Name ' values
df_trend[' Month Name '] = df_trend[' Month Name '].str.strip()

# Create a new column 'Sort Order' for sorting
df_trend['Sort Order'] = df_trend.apply(lambda row: row['Year'] * 100 + month_order.index(row[' Month Name ']), axis=1)

# Sort the DataFrame based on the custom sort order
df_trend = df_trend.sort_values('Sort Order')

# Drop the 'Sort Order' column
df_trend = df_trend.drop(columns='Sort Order')

# Combine 'Year' and ' Month Name ' columns for x-axis labels
df_trend['Month Year'] = df_trend[' Month Name '] + ' ' + df_trend['Year'].astype(str)

# Set the figure size
plt.figure(figsize=(15, 6))

# Create the line plot
plt.plot(df_trend['Month Year'], df_trend['  Sales '], color='#b80045')
plt.xticks(rotation='vertical', size=8)

# Add labels and title
plt.xlabel('Month and Year')
plt.ylabel('Total Sales')
plt.title('Sales Trend Over Months and Years')

# Show the plot
plt.tight_layout()
plt.show()
```

- <h2>  which are the top 5 products by sales?

```python
# Group by 'Product' and sum '  Sales '
product_sales = df.groupby(' Product ')[['  Sales ']].sum()

# Sort the results in descending order and select the top 5
top_5_products = product_sales.sort_values(by='  Sales ', ascending=False).head(5)

# Print the top 5 products by sales
top_5_products
```

```python
# Create a bar plot for the top 5 products by sales
plt.figure(figsize=(10, 6))
plt.bar(top_5_products.index, top_5_products['  Sales '], color='blue')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Top 5 Products by Sales')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

- <h4> which are the most selling product by quantity

```python
# Group by 'Product' and sum 'Units Sold'
most_selling_products = df.groupby(' Product ')[['Units Sold']].sum()

# Sort the results in descending order and select the most selling product
most_selling_product = most_selling_products.sort_values(by='Units Sold', ascending=False).head(1)

# Print the most selling product
print("Most Selling Product by Quantity:")
most_selling_product
```
